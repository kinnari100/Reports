```python
import fitz
import glob, os
import pandas as pd
import openpyxl

DIR='/Users/kinnaripatel/Desktop/k/kpdf/fwdabstractreportrequestpleaseadvise'
os.chdir(DIR)

def listToString(s):  
    str1 = ""  
    for ele in s:  
        str1 += ele   
    return str1  

for file in glob.glob("*.pdf"):
    print(file)
    filename = os.path.splitext(file)
    filename = filename[0]
    pdfs = []

    with fitz.open(file) as doc:
        text = ""
        for page in doc:
            text += page.get_text(text)
            pdfs.append(text)
        
        #textfile = open(DIR + 'text\\' + filename + '.txt', 'w',encoding="utf-8")
        #pages = listToString(pdfs)
        #textfile.write(pages)
        #textfile.close()

    data = text

    df = pd.DataFrame([x.split(';') for x in data.split('\n')[1:]], columns=[x for x in data.split('\n')[0].split(';')])
    #print(df)
    df=df.rename(index={0:'text'})
    df1=df.transpose()


    df2 = df1[df1.columns[0:]].apply(lambda x: ' '.join(x.dropna().astype(str)),axis=1)

    with pd.ExcelWriter('/Users/kinnaripatel/Desktop/k/kpdf/kp/kfinal.xlsx', engine='openpyxl',mode='a',if_sheet_exists='overlay') as writer:
            pd.DataFrame(df2).to_excel(writer,sheet_name=filename,index=False)
    print(df2)
    
    xls = pd.ExcelFile('/Users/kinnaripatel/Desktop/k/kpdf/kp/kfinal.xlsx')
    df_main = pd.DataFrame()
    for sheet in xls.sheet_names:
        df = pd.read_excel(xls, sheet)
        df['Sheet-Name'] = sheet
        #df_main = df_main.append(df)
        df_main = pd.concat([df_main, df], ignore_index = True)
        #print(df_main)
        
df_main.to_excel ("/Users/kinnaripatel/Desktop/k/kpdf/kp/kfinal2.xlsx", index = False, header=True)
```
